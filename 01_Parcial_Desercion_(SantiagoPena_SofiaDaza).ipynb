# 📺 Deserción de estudiantes

## Contexto y Variables Principales

Los datos utilizados corresponden a un estudio sobre deserción y éxito académico en educación superior. Cada registro representa a un estudiante de una institución de educación superior, integrando información obtenida de varias bases de datos institucionales. El conjunto incluye 3,630 estudiantes y 36 variables predictoras, con información disponible desde el momento de la matrícula y también sobre el desempeño académico en los primeros semestres.

Se dispone de variables relacionadas con el perfil académico previo 🎓 (tipo de formación anterior, nota de calificación previa, orden y modalidad de aplicación, programa al que ingresa), características demográficas y personales 🌍 (nacionalidad, estado civil, asistencia diurna/nocturna), así como factores familiares y socioeconómicos 👨‍👩‍👧‍👦💼 (nivel educativo y ocupación de la madre y del padre, condición de deudor, beca, pago de matrícula al día). También se incluyen indicadores del desempeño universitario temprano 📚 (créditos cursados, aprobados, evaluaciones y calificaciones en el primer y segundo semestre), que permiten capturar señales iniciales del progreso académico del estudiante.

El objetivo principal es predecir la variable de respuesta Target, que clasifica a los estudiantes en tres categorías: Dropout y Graduate. Este problema se plantea como una tarea de clasificación, con desbalance entre categorías, y resulta especialmente relevante porque permite identificar de manera temprana a estudiantes en riesgo de deserción ⚠️, apoyar la toma de decisiones institucionales 🏫 y diseñar estrategias de acompañamiento más efectivas para mejorar la permanencia y el éxito académico.
