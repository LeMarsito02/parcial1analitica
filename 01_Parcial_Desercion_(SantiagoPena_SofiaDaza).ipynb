## Contexto y Variables Principales

Los datos utilizados corresponden a un estudio sobre deserción y éxito académico en educación superior. Cada registro representa a un estudiante de una institución de educación superior, integrando información obtenida de varias bases de datos institucionales. El conjunto incluye 3,630 estudiantes y 36 variables predictoras, con información disponible desde el momento de la matrícula y también sobre el desempeño académico en los primeros semestres.

Se dispone de variables relacionadas con el perfil académico previo 🎓 (tipo de formación anterior, nota de calificación previa, orden y modalidad de aplicación, programa al que ingresa), características demográficas y personales 🌍 (nacionalidad, estado civil, asistencia diurna/nocturna), así como factores familiares y socioeconómicos 👨‍👩‍👧‍👦💼 (nivel educativo y ocupación de la madre y del padre, condición de deudor, beca, pago de matrícula al día). El objetivo principal es predecir la variable de respuesta Target, que clasifica a los estudiantes en tres categorías: Dropout y Graduate. Este problema se plantea como una tarea de clasificación, con desbalance entre categorías, y resulta especialmente relevante porque permite identificar de manera temprana a estudiantes en riesgo de deserción ⚠️, apoyar la toma de decisiones institucionales 🏫 y diseñar estrategias de acompañamiento más efectivas para mejorar la permanencia y el éxito académico.

## Datos Atipicos (Outliers)

### Preparacion de quantiles

```python
# por si tienes el tab en el nombre
df = df.rename(columns={"Daytime/evening attendance\t": "Daytime/evening attendance"})

outlier_summary = []
outlier_examples = {}  # guarda un ejemplo de filas outlier por columna

for col in numeric_cols:
    s = pd.to_numeric(df[col], errors="coerce")  # asegura numérico

    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mask_out = (s < lower) | (s > upper)
    n_out = int(mask_out.sum())
    pct_out = (n_out / len(df)) * 100

    outlier_summary.append({
        "variable": col,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "lower": lower,
        "upper": upper,
        "n_outliers": n_out,
        "pct_outliers": pct_out
    })

    # guarda hasta 5 filas outlier como ejemplo (para documentar)
    outlier_examples[col] = df.loc[mask_out, [col, "Target", "Target_encoded"]].head(5)

summary_df = pd.DataFrame(outlier_summary).sort_values("n_outliers", ascending=False)

# tabla resumen final
summary_df
```

## Tratamiento de Outliers

```python
bounds = {}

for col in numeric_cols:
    s = pd.to_numeric(df[col], errors="coerce")

    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # guardamos límites para el reporte
    bounds[col] = (lower, upper)

    # tratamiento
    df[col] = s.clip(lower, upper)

print("Outliers tratados usando Winsorization.")
```

#### Para el tratamiento de los outliers se uso el metodo de winsorización que utiliza el rango entre los cuantiles Q1 y Q3 para poder reducir la influencia de valores lo que ayuda a que el modelo sea mas acertado al no eliminar observaciones. Lo que hace es reducir el impacto de valores extremos remplazandolos con valores menos extremos basandose en el rango de los percentiles.

```python
check_outliers = []

for col in numeric_cols:
    s = df[col]

    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((s < lower) | (s > upper)).sum()

    check_outliers.append({
        "variable": col,
        "outliers_post_tratamiento": outliers
    })

pd.DataFrame(check_outliers)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import numpy as np
import matplotlib.pyplot as plt

# clases del modelo
classes_ = model.named_steps["clf"].classes_

# Probabilidad de la clase positiva
y_score = y_proba[:, 1]

# Convertimos y_test a binaria (1 = clase positiva)
y_test_bin = (y_test == classes_[1]).astype(int)

# Calcular ROC
fpr, tpr, thresholds = roc_curve(y_test_bin, y_score)
auc = roc_auc_score(y_test_bin, y_score)

# Mejor umbral usando Youden's J
youden = tpr - fpr
best_idx = np.argmax(youden)

best_threshold = thresholds[best_idx]
best_fpr = fpr[best_idx]
best_tpr = tpr[best_idx]

print(f"Mejor umbral (Youden): {best_threshold:.4f}")
print(f"TPR: {best_tpr:.3f}")
print(f"FPR: {best_fpr:.3f}")
print(f"AUC: {auc:.3f}")

# -----------------------------
# Plot ROC con el puntico
# -----------------------------
plt.figure(figsize=(6,5))

plt.plot(fpr, tpr, label=f"ROC (AUC = {auc:.3f})")
plt.plot([0,1], [0,1], linestyle="--", label="Azar")

# punto del mejor umbral
plt.scatter(best_fpr, best_tpr,
            color="red",
            s=80,
            label=f"Mejor umbral = {best_threshold:.3f}")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve con mejor umbral")
plt.legend()

plt.tight_layout()
plt.show()
```
